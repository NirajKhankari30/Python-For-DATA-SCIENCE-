# While loop

A while loop in Python is used to repeatedly execute a block of code as long as a certain condition is true. It allows you to create a loop that continues until the condition becomes false.

```python
# Initialize the loop counter to 1
i=1
while i<=5:

    # Print the value of i
    print(i)

    # Increment the value of i by 1 in each iteration
    i=i+1
```

In this code, we initialize the variable i to 1. The while loop checks the condition i <= 5. If the condition is true, the code inside the loop is executed, which includes printing the value of i and incrementing i by 1. This process continues until i becomes 6, at which point the condition becomes false, and the control flows to the else block.

```python
i=1
while i<=5:
    # Print the value of i
    print(i)
    # Increment the value of i by 1
    i=i+1
else:
     # Print a message when the condition becomes false
    print("i is no longer less than 6")
```
